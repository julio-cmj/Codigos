## Visualizar os resultados de maneiras diferentes

```python
from math import pi
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, DensityMatrix
 
qc = QuantumCircuit(2)
qc.h(0)
qc.crx(pi / 2, 0, 1)
psi = Statevector(qc)

print(psi)
```

```python
DensityMatrix(psi).draw('latex')
```

```python
psi.draw('latex')
```

```python
psi.draw('city')
```

```python
psi.draw('bloch')
```
